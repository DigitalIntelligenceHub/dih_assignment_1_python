```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib
```

```python
# refrence:  http://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_csv.html#pandas.read_csv
df=pd.read_csv("/resources/data/chronic_kidney_disease_updated.csv", sep=",")
```

```python
# first five rows in dataset
df.head(5)
```

```python
# names of columns
print df.columns.values
```

```python
# refrence:  http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.unique.html
# by checking the type of values in dm variable, it shows that data set is not clean
print df.dm.unique()
```

```python
# refrence: http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.replace.html#pandas.DataFrame.replace
df["dm"].replace(regex=True,inplace=True, to_replace=r'\t',value=r'')
```

```python
# refrence: http://stackoverflow.com/questions/17097236/how-to-replace-values-with-none-in-pandas-data-frame-in-python
df.replace('0',np.nan)
```

```python
# to see the removed anomalies
print df.dm.unique()
```

```python
# refrence: http://stackoverflow.com/questions/17097236/how-to-replace-values-with-none-in-pandas-data-frame-in-python
# remove '?' from data frame
df=df.replace('\?',np.nan, regex=True)
```

```python
print df.head()
```

```python
print df.dm.unique()
```

```python
# change data type to numerical
# refrence: http://stackoverflow.com/questions/15891038/pandas-change-data-type-of-columns
df[['age','bp','bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']]= df[['age','bp','bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']].apply(pd.to_numeric, errors='ignore')
```

```python
# to see the changed data types
print df.dtypes
```

```python
# maximum blood pressure
print df['bp'].max()
```

```python
# refrence: http://stackoverflow.com/questions/19392226/attributeerror-dataframe-object-has-no-attribute
# Count the number of normal or abnormal red blood cells people having CKD
df['rbc'].value_counts()
```

```python
# refrence: http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.to_csv.html#pandas.DataFrame.to_csv
# csv file of cleaned data
df.to_csv("/resources/data/chronic_kidney_disease_updated.csv")
```
